```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(x):
    """The function we want to minimize: f(x) = (x+3)²"""
    return (x + 3)**2
```

```python
def derivative(x):
    """Derivative of the function: f'(x) = 2(x+3)"""
    return 2 * (x + 3)
```

```python
def gradient_descent(start_x, learning_rate=0.1, num_iterations=100, tolerance=1e-6):
    """
    Performs gradient descent optimization.

    Parameters:
    - start_x: Starting point
    - learning_rate: Step size for each iteration
    - num_iterations: Maximum number of iterations
    - tolerance: Minimum change in x to continue iteration

    Returns:
    - x_history: List of x values during optimization
    - y_history: List of corresponding function values
    """
    x = start_x
    x_history = [x]
    y_history = [function(x)]

    for i in range(num_iterations):
        grad = derivative(x)
        new_x = x - learning_rate * grad

        # Check if we've converged
        if abs(new_x - x) < tolerance:
            break

        x = new_x
        x_history.append(x)
        y_history.append(function(x))

    return x_history, y_history
```

```python
# Run gradient descent
start_x = 2
x_history, y_history = gradient_descent(start_x)
```

```python
# Create points for plotting the function
x_plot = np.linspace(-5, 5, 100)
y_plot = [function(x) for x in x_plot]
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x_plot, y_plot, 'b-', label='f(x) = (x+3)²')
plt.plot(x_history, y_history, 'ro-', label='Gradient descent path')
plt.scatter(x_history[-1], y_history[-1], color='green', s=100, label='Final minimum')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent Optimization')
plt.legend()
plt.show()
```

```python
# Print results
print(f"Starting point: x = {start_x}, f(x) = {function(start_x):.4f}")
print(f"Minimum found at: x = {x_history[-1]:.4f}, f(x) = {y_history[-1]:.4f}")
print(f"Number of iterations: {len(x_history)}")
```
